# Reuters Dataset

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import reuters
from tensorflow.keras import models, layers
from tensorflow.keras.utils import to_categorical
```

```python
(num_words, max_len) = (10000, None)
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=num_words)
```

```python
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.0
    return results

x_train = vectorize_sequences(x_train)
x_test = vectorize_sequences(x_test)
```

```python
# One-hot encode labels
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(64, activation='relu', kernel_initializer="he_normal"))
model.add(layers.Dense(46, activation='softmax'))
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy',  metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=8, batch_size=512, validation_split=0.2)
```

```python
results = model.evaluate(x_test, y_test)
print(f"test case accuracy {results[1]}")
```
